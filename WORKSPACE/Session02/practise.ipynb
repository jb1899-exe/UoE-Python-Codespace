```python
while True:
    lucky_input = input("Please choose a number between 1 and 120: ")
    if lucky_input.isdigit():
        lucky_number = int(lucky_input)
        if 1 <= lucky_number <= 120:
            print(f"Thank you! You entered: {lucky_number}")
            break
    print("Invalid input. Try again.")
```

```python
def display_info(**info):

    for key, value in info.items():

        print(f"{key}: {value}")



display_info(name="Alice", age=25, city="London")
```

```python
contacts = dict()
while True:
    name_response = input("Please enter your name (enter 'quit' to exit): ")
    if name_response.lower() in ['quit', 'q']:
        print("Quitting program...")
        break
    else:
        while True:
            num_response = input(f"Hello {name_response}!, please enter your favourite number (enter 'quit' to exit): ")
            if num_response.lower() in ['quit', 'q']:
                break
            else:
                contacts.update({name_response: num_response})
                print(f"Saving contact '{name_response}' with number '{num_response}'")
                break

for name, num in contacts.items():
    print(f"Name: {name}, Num: {num}")
```
